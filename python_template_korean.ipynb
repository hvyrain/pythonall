# 예제 : 사인-코사인 그래프

```python
# 샘플 그래프 코드
import numpy as np
i = np.linspace(-np.pi, np.pi, 60)
s = np.sin(i)
c = np.cos(i)
plt.figure(figsize=(8, 5))
#plt.style.use('ggplot')

plt.plot(i, s, 'r-', label='사인')
plt.plot(i, c, 'b--', label='코사인')
plt.title('사인-코사인 그래프')
plt.legend()
plt.grid(True)
plt.show()
```
